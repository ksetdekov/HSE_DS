запросы запускаются через sqlite из питона, пути подразумевают, что файл с базой данных находится там же, где и этот ноутбук

импорт библиотек и код для выполнения запросов и вывода их

```python
import sqlite3
import pandas as pd 


def get_data(sql_query):
    """function to get as a pandas dataframe data from a query from
    footbal data"""
    
    conn = sqlite3.connect('database.sqlite')  # footbal db name
    return pd.read_sql_query(sql_query, conn)
    conn.close()
```

## 1
Объедините таблицы Country, League. Должно выводится на экран 5 колонок (2 из Country, 3 из League)

```python
query = """
SELECT * FROM Country JOIN League on Country.id = League.id;
"""
get_data(query)
```

## 2
Напишите запрос, который сопоставит country_id, league_id их названиям (страны/лиги) в таблице Match. Выведите колонки country_id, название страны, league_id, название лиги, date (всего 5 колонок). Ограничьте запрос 50 значениями.

```python
query = """
SELECT Match.country_id, Country.name, Match.league_id, League.name, date 
FROM Match 
LEFT JOIN Country on Country.id = Match.country_id 
LEFT JOIN League on League.id = Match.league_id 
LIMIT 50;
"""
get_data(query)
```

## 3
Подсчитайте сколько значений будет содержаться в декартовом произведении таблиц Country и Player. Выведите число.

```python
query = """
SELECT count(*) as id_count
FROM Country 
CROSS JOIN Player;
"""
get_data(query)['id_count'][0]
```

## 4
Определите кол-во футболистов, overall_rating которых когда-либо был больше, чем у игрока с player_api_id равным 164684 в день 2016-02-18 (таблица Player_Attributes). Сделайте это без использования подзапросов.

```python
query = """
SELECT count(*) as id_count
FROM Player_Attributes AS p1,  Player_Attributes AS p2
WHERE p2.player_api_id = 164684 AND p2.date = '2016-02-18 00:00:00'
AND p1.overall_rating > p2.overall_rating;
"""
get_data(query)['id_count'][0]
```

## 5
Определите средний рост футболистов, вес которых больше среднего значения по всем игрокам (через подзапрос).

```python
query = """
SELECT avg(height) as avg_height FROM Player 
WHERE weight/2.205 > (
SELECT avg(weight/2.205) as avg_weight FROM Player
);
"""
get_data(query)['avg_height'][0]
```

## 6
Сколько строк получится, если применить LEFT JOIN таблиц Country и Match по колонке country_id?


```python
query = """
SELECT count(*) as id_count
FROM Country 
LEFT JOIN Match on Country.id = Match.country_id;
"""
get_data(query)['id_count'][0]
```

##  Бонус 1
Выведите все имена и фамилии (player_name) футболистов, у которых вес больше среднего веса всех футболистов без использования подзапроса (необходимо использовать cross join или self join). 

```python
query = """
SELECT *
FROM (
  SELECT player_name, weight AS weight, avg(weight) OVER () AS avg_weight
  FROM Player
) ttable
WHERE weight > avg_weight
"""
get_data(query)
```

## Бонус 2
Удалите дубликаты из запроса со слайда № 10. (5б.)


поскольку дубликаты будут двумя идентичными строками в разном порядке, на них сработает сравнение, чтобы выкинуть гарантированно только одно сравнение

```python
query = """
SELECT p1.player_name, p2.player_name
FROM Player AS p1, Player AS p2
WHERE p1.birthday = p2.birthday AND p1.id <> p2.id 
AND p1.player_name < p2.player_name;
"""
get_data(query)
```
