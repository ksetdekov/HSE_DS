## Overview

This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. [DBFS](https://docs.databricks.com/user-guide/dbfs-databricks-file-system.html) is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in **Python** so the default cell type is Python. However, you can use different languages by using the `%LANGUAGE` syntax. Python, Scala, SQL, and R are all supported.

```
# File location and type
file_calendar = "/FileStore/tables/calendar.csv"
file_validation = "/FileStore/tables/sales_train_validation.csv"
file_evaluation = "/FileStore/tables/sales_train_evaluation.csv"
file_prices = "/FileStore/tables/sell_prices.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df_calendar = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_calendar)

df_validation = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_validation)

df_evaluation = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_evaluation)

df_prices = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_prices)

display(df_prices)
```

```
from pyspark.sql.window import Window
import pyspark.sql.functions  as F

window_lag = Window.partitionBy("id").orderBy("date")
window_agg = Window.partitionBy("id").orderBy("date").rangeBetween(Window.unboundedPreceding, Window.currentRow)
window_lag_store =  Window.partitionBy("store_id").orderBy("date")
window_lag_store_category =  Window.partitionBy(["store_id", "state_id"])

result = (
  result.withColumn("lag_1", F.lag('sales', 1).over(window_lag))
  .withColumn("store_lag_1", F.lag('sales', 1).over(window_lag_store))
  .withColumn("mean", F.mean("lag_1").over(window_agg))
)
```
